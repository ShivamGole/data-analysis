```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Read the Dataset

```python
df = pd.read_csv("/content/sales.csv",encoding = "ISO-8859-1")
df
```

Check for Null Data

```python
df["Amount"].isnull().sum()
```

delete unwanted columns

```python
del df['Status']
del df['unnamed1']
```

handel or delete a nana value

```python
#df.fillna(method='ffill', inplace=True)
df.dropna(inplace=True)
```

```python
#df.reset_index(drop=True, inplace=True)
```

Rename column name

```python
df.rename(columns={"Marital_Status": "Marital"},inplace=True)
```

modify column value

```python
df['Marital']=df['Marital'].map({0: 'Single', 1:'Married'})

df['Gender'] = df['Gender'].map({'F': 'Female', 'M': 'Male'})
```

check a visulaization

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

group by

```python
#gender
print(df.groupby('Gender')['Amount'].sum().sort_values(ascending=False))

# Age Group
print(df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False))

# State
print(df.groupby('State')['Amount'].sum().sort_values(ascending=False))

# Occupation
print(df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False))

# Product Category
print(df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False))

# Marital Status
print(df.groupby('Marital')['Amount'].sum().sort_values(ascending=False))
```

```python
sns.barplot(x='Gender', y='Amount', data=df, estimator=sum)
plt.title("Total Spending by Gender")
plt.show()
```

```python
# 🧑‍🤝‍🧑 Gender-Based Spending

# The [Gender] group spent the most, indicating a higher purchasing tendency among [Gender] customers.

# 📅 Age Group Spending

# The [Age Group] category contributed the most to the total amount spent, showing that this group is the most active in shopping.

# 📍 State-wise Contribution

# [State] had the highest overall spending, making it the top-performing region in the dataset.

# 💼 Occupation Insights

# People working in [Occupation] spent the most, suggesting a strong correlation between this profession and purchasing power.

# 📦 Product Category Performance

# The [Product Category] category earned the most revenue, making it the most popular or expensive product group.

# 💍 Marital Status Comparison

# [Married/Single] individuals showed higher spending patterns compared to their counterparts.
```
